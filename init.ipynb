```python

def features_significance(features, target, df):

base = {}
for feature in features:

  df_anali = pd.crosstab(df[target], df[feature])


  stat, p, dof, expected = chi2_contingency(df_anali)

  prob = 0.95

  critical = chi2.ppf(prob, dof)
  alpha = 1.0 - prob


  if abs(stat) >= critical:
    pass
  else:
    if p <= alpha:
      pass
    else:
      base[feature] = p

return base
```
